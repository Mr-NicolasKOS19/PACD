```python
import pandas as pd
import matplotlib.pyplot as plt

#Criando o conjunto de dados
dados = {
    'cidade' : ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Salvador', 'Manaus'],
    'Temperatura' : [ 28.5, 34.2, 22.0, 31.8, 33.5],
    'umidade' : [ 60, 75, 55, 80, 90]
}

df = pd.DataFrame(dados)
print ("Dados Carregados com Sucesso!")
df
```

```python
#Calculando Métricas Importantes
temp_media = df['Temperatura'].mean()
cidade_quente = df.loc[df['Temperatura'].idxmax(), 'cidade']
cidade_fria = df.loc[df['Temperatura'].idxmin(), 'cidade']

print (f"A Temperatura média geral é: {temp_media:.2f} ºC")
print (f"A Cidade mais quente é: {cidade_quente}")
print (f"A Cidade mais fria é: {cidade_fria}")
```

```python
# Criando Grafico
plt.figure(figsize=(10, 6))
plt.bar (df['cidade'], df['Temperatura'], color=['skyblue', 'orange', 'purple', 'red', 'darkred'])

#Customizaçao
plt.title('Comparativo de cores por cidade', fontsize=14)
plt.xlabel('cidades', fontsize=16)
plt.ylabel('Temperatura (ºC)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

#Exibir
plt.show()
```
